```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("M:\Pat_Mini_Project\data\Final DF\Final_Mobile_data.csv")

data
```

```python
data.drop(columns=['Unnamed: 0'], inplace=True)

data['Price'] = data['Price'].str.replace('Rs.', '').str.replace(',', '')
```

```python
data
```

```python
data['RAM'] = data['RAM'].str.replace('GB', '').str.replace('RAM', '').str.replace(' ', '')
data['Refresh Rate'] = data['Refresh Rate'].str.replace('Hz Refresh Rate', '').str.replace(' ', '')
data['Battery'] = data['Battery'].str.replace(' ', '').str.replace('mAh', '')

data
```

```python
data['Price'] = data['Price'].astype(float)
data['Battery'] = data['Battery'].astype(float)
data['RAM'] = data['RAM'].astype(float)
data['Refresh Rate'] = data['Refresh Rate'].astype(float)
```

```python
# Encode categorical variables
encoder = OneHotEncoder()
categorical_cols = ['Brand', 'os', 'OS', 'Core', 'Processor', 'Display Size', 'Display Type', 'Camera Details', 'Flash Type', 'Screen flash', 'Fast charging', 'Type-C']
encoded_data = encoder.fit_transform(data[categorical_cols]).toarray()
encoded_cols = pd.DataFrame(encoded_data, columns=[f'{col}_{val}' for col in categorical_cols for val in encoder.categories_[categorical_cols.index(col)]])

# Concatenate encoded columns with the original data
data = pd.concat([data, encoded_cols], axis=1)

# Normalize numerical features
scaler = StandardScaler()
numerical_cols = ['Price', 'Battery', 'RAM', 'Refresh Rate']
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Continue with the rest of your code (e.g., similarity calculation, recommendation generation)
```

```python
# Select features for similarity calculation
features = data[['Price', 'Battery', 'RAM', 'Refresh Rate'] + list(encoded_cols.columns)]

# Calculate cosine similarity matrix
similarity_matrix = cosine_similarity(features, features)
```

```python
def get_recommendations(model, data, features, selected_features, top_n=5):
    # Filter data based on selected features
    selected_data = data.copy()
    print("Selected Features:", selected_features)
    for feature, value in selected_features.items():
        if feature in numerical_cols:
            selected_data = selected_data[selected_data[feature].round(2) == round(value, 2)]
        else:
            selected_data = selected_data[selected_data[feature + '_' + str(value)] == 1]

    # Check if filtered data has at least one sample
    if len(selected_data) == 0:
        print("No matching smartphones found.")
        return None  # or return an empty DataFrame or raise an exception

    # Calculate similarity with selected smartphones
    selected_features_values = selected_data[features].values
    similarity_scores = cosine_similarity(selected_features_values, data[features])

    # Get indices of top similar smartphones
    top_indices = similarity_scores.argsort()[0][-top_n-1:-1][::-1]

    # Get top recommendations
    recommendations = data.iloc[top_indices][["Final Rating"]]
    
    return recommendations
```

```python

# Example: User selects desired features
selected_features = {'Price': 17499}

# Get recommendations
recommendations = get_recommendations(cosine_similarity, data, ['Price', 'Battery', 'RAM', 'Refresh Rate'] + list(encoded_cols.columns), selected_features)
print(recommendations)
```
