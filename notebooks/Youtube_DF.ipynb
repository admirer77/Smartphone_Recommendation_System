```python
# Define search parameters
query = 'Prasad Tech in Telugu single mobile review launched in 2024'
max_results = 50  # Adjust as needed
order = 'relevance'  # Other options: 'rating', 'viewCount', 'date', etc.
```

```python

search_response = youtube.search().list(
    q=query,
    part='snippet',
    type='video',
    maxResults=max_results,
    order=order
).execute()
```

```python

video_ids = [item['id']['videoId'] for item in search_response['items']]

videos = []
```

```python
for video_id in video_ids:
    video_response = youtube.videos().list(
        part='snippet,statistics',
        id=video_id
    ).execute()

    video_details = video_response['items'][0]
    try:
        dislikes = video_details['statistics']['dislikeCount']
    except KeyError:
        dislikes = 0  # Set default value if dislikeCount is missing

    videos.append({
        'Title': video_details['snippet']['title'],
        'Video ID': video_id,
        'Channel': video_details['snippet']['channelTitle'],
        'Published At': video_details['snippet']['publishedAt'],
        'Views': video_details['statistics']['viewCount'],
        'Likes': video_details['statistics']['likeCount'],
    })
```

```python
videos_df = pd.DataFrame(videos)
videos_df
```
