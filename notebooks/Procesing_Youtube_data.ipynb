```python
import pandas as pd
```

```python
Android_df = pd.read_csv("M:\Pat_Mini_Project\data\Youtube\youtube_android_review.csv")

Android_df
```

```python
Android_df = Android_df[~Android_df['Title'].str.contains('tech news', case=False)]
Android_df = Android_df[~Android_df['Title'].str.contains('technews', case=False)]
Android_df
```

```python
Android_df = Android_df[Android_df['Channel'] == 'Prasadtechintelugu']

Android_df
```

```python
Android_df["Youtube Rating"] = Android_df["Likes"] / Android_df["Views"]

Android_df
```

```python
Android_df["Youtube Rating"].max()
```

```python
c = 10 / Android_df["Youtube Rating"].max()

c
```

```python
Android_df["Youtube Rating"] = Android_df["Youtube Rating"] * c

Android_df
```

```python
Android_df = Android_df.reset_index(drop=True)

Android_df
```

```python
Android_df = Android_df[["Title", "Video ID", "Channel", "Published At", "Views", "Likes", "smartphone Name", "Youtube Rating"]]

Android_df
```

```python
Android_df_new = Android_df.groupby('smartphone Name')['Youtube Rating'].mean().reset_index()

Android_df_new
```

```python
Android_df_new["Youtube Rating"].max()
```

```python
c2 = 10 - Android_df_new["Youtube Rating"].max()

c2
```

```python
Android_df_new["Youtube Rating"] = Android_df_new["Youtube Rating"] + c1

Android_df_new
```

```python
Android_df_new.to_csv("Processed_YT_Android_Review.csv")
```
