```python
import pandas as pd
```

```python
android_df1 = pd.read_csv("M:\Pat_Mini_Project\data\OS\Android_mobiles.csv")

android_df1.drop(columns=['Unnamed: 0'], inplace=True)
android_df1
```

```python
android_df2 = pd.read_csv("M:\Pat_Mini_Project\data\Youtube\Processed_YT_Android_Review.csv")

android_df2["Name"] = android_df2["smartphone Name"]

android_df2 = android_df2[["Name", "Youtube Rating"]]

android_df2
```

```python
android_df = pd.merge(android_df1,android_df2, on="Name")

android_df
```

```python
android_df["Final Rating"] = (android_df["91 mobiles rating"] * 0.7) + (android_df["Youtube Rating"] * 0.3)

android_df
```
