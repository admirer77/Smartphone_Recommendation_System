```python
import pandas as pd
```

```python
Apple_df = pd.read_csv("..\data\Apple.csv")

Apple_df
```

```python
Apple_df = Apple_df.drop_duplicates()

Apple_df
```

```python
Oneplus_df = pd.read_csv("..\data\Oneplus.csv")

Oneplus_df
```

```python
Oneplus_df = Oneplus_df.drop_duplicates()

Oneplus_df
```

```python
Realme_df = pd.read_csv("..\data\Realme.csv")

Realme_df
```

```python
Realme_df = Realme_df.drop_duplicates()

Realme_df
```

```python
Realme_df['Price'].fillna("Rs.10,499", inplace=True)

Realme_df
```

```python
Redme_df = pd.read_csv("..\data\Redmi.csv")

Redme_df
```

```python
Redme_df = Redme_df.drop_duplicates()

Redme_df
```

```python
Samsung_df = pd.read_csv("..\data\Samsung.csv")

Samsung_df
```

```python
Samsung_df = Samsung_df.drop_duplicates()

Samsung_df
```

```python
Vivo_df = pd.read_csv("..\data\Vivo.csv")

Vivo_df
```

```python
Vivo_df = Vivo_df.drop_duplicates()
```

```python
Android_df = pd.concat([Oneplus_df, Realme_df, Redme_df, Samsung_df, Vivo_df], ignore_index=True)

# If you want to reset the index of the combined dataframe
Android_df.reset_index(drop=True, inplace=True)

Android_df
```

```python
Android_df['Price'] = Android_df['Price'].str.replace('Rs.', '').str.replace(',', '')
```

```python
Android_df
```

```python
Android_df['91 mobiles rating'] = Android_df['91 mobiles rating'].str.replace('%', '')
```

```python
Android_df['91 mobiles rating'] = pd.to_numeric(Android_df['91 mobiles rating']) / 10
```

```python
Android_df
```

```python
Android_df_1 = Android_df['Specs'].str.split('\n', expand=True)

column_names = ['Core','Processor', 'RAM','Empty3', 'Display Size', 'Display Type', 'Refresh Rate','Empty', 'Camera Details', 'Flash Type','Screen flash','Empty2','Battery','Fast charging', 'Type-C']
Android_df_1.columns = column_names
Android_df_1 = Android_df_1.drop(columns=['Empty3','Empty','Empty2'])

Android_df_1
```

```python
Android_df = pd.concat([Android_df, Android_df_1], axis=1)

Android_df
```

```python
Android_df = Android_df.drop(columns=['Specs'])

Android_df
```

```python
Android_df.to_csv("Android_mobiles.csv")
```
